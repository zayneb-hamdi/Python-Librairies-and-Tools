```python
plt.figure(1)
plt.pie(x = [8000,20000,16200], labels = ['math','sciences','economie'],
           colors = ["#f33030","#218E12","#1448E3"],
           autopct = lambda x: str(round(x, 2)) + '%')
plt.legend()
```

```python
plt.figure(2)
zayneb = [10,20,26,31,12,48,19]
mariem = [12,25,30,33,16,50,22]
plt.bar(range(len(zayneb)), zayneb, label = "zayneb 's sales",color ="#430651" )
plt.bar(range(len(mariem)), mariem, bottom = zayneb, label = "mariem's sales",color ="#e084e3" )
plt.xticks([0,1,2,3,4,5,6], ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat','Sun'])
plt.title("The sales based on week days")
plt.legend()
```
